```python
def clean_headline(text):
    if pd.isna(text):
        return ""
    text = str(text)
    text = re.sub(r"^b['\"]", "", text)
    text = re.sub(r"['\"]$", "", text)
    text = text.replace("&nbsp;", " ")
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
from scipy import stats

sentiment_up = df[df['Label'] == 1]['sentiment']
sentiment_down = df[df['Label'] == 0]['sentiment']

t_stat, p_value = stats.ttest_ind(sentiment_up, sentiment_down)
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Signifikan secara statistik (p < 0.05)")
else:
    print("Not statistically significant (p >= 0.05)")
```

```python
from collections import Counter
import re


stopwords = set(['the', 'a', 'an', 'to', 'of', 'in', 'and', 'for', 'on', 'is', 
                  'that', 'with', 'as', 'at', 'by', 'from', 'has', 'have', 'was',
                  'were', 'be', 'been', 'are', 'it', 'this', 'will', 'not', 'but',
                  'their', 'they', 'after', 'says', 'said', 'over', 'new', 'us',
                  'up', 'out', 'who', 'his', 'her', 'its', 'we', 'you', 'i'])

def get_words(text):
    words = re.findall(r'\b[a-z]{3,}\b', text.lower())  # ambil kata min 3 huruf
    return [w for w in words if w not in stopwords]


text_down = ' '.join(df[df['Label'] == 0]['all_headlines'])
text_up = ' '.join(df[df['Label'] == 1]['all_headlines'])

words_down = get_words(text_down)
words_up = get_words(text_up)

# Top 20 kata paling sering di masing-masing kondisi
print("=== Top 20 Words – Market Downturn Day ===")
print(Counter(words_down).most_common(20))

print("\n=== TOP 20 WORDS - BULL MARKET DAY ===")
print(Counter(words_up).most_common(20))
```

```python
from collections import Counter

count_down = Counter(words_down)
count_up = Counter(words_up)

total_down = len(words_down)
total_up = len(words_up)


all_candidate_words = set([w for w, c in count_down.items() if c >= 15] + 
                            [w for w, c in count_up.items() if c >= 15])


ratios = []
for word in all_candidate_words:
    freq_down = count_down.get(word, 0) / total_down
    freq_up = count_up.get(word, 0) / total_up
    ratio = freq_down / (freq_up + 0.00001)  # hindari pembagian nol
    ratios.append((word, ratio, count_down.get(word,0), count_up.get(word,0)))


ratios_sorted_down = sorted(ratios, key=lambda x: -x[1])
print("=== THE MOST TYPICAL WORDS ON MARKET DOWN DAYS ===")
for word, ratio, cd, cu in ratios_sorted_down[:15]:
    print(f"{word}: ratio={ratio:.2f}, count_down={cd}, count_up={cu}")


ratios_sorted_up = sorted(ratios, key=lambda x: x[1])
print("\n=== THE MOST TYPICAL WORDS ON A DAY THE MARKET IS UP ===")
for word, ratio, cd, cu in ratios_sorted_up[:15]:
    print(f"{word}: ratio={ratio:.2f}, count_down={cd}, count_up={cu}")
```
